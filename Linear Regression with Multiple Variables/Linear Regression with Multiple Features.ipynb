```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error as MAE
```

```python
class LinReg:
    def __init__(self,iterations,rate):
        self.iter=iterations
        self.rate=rate
    def param_vec(self,train_x):
        self.param=np.random.rand(train_x.shape[1],1)
        #print(self.param)
    def cost_function(self,train_y):
        return np.nansum(np.square(self.pred_y-train_y))/(2*train_y.shape[0])
    def predict(self,train_x):
        self.pred_y=np.dot(train_x,self.param)
    def descent(self,train_x,train_y):
        x=self.pred_y-train_y   # n,1
        x=np.dot(train_x.T,x)
        x=x*self.rate
        x=x/train_y.shape[0]
        #print(x.shape)
        self.param=self.param-x
        
    def normalizex(self,train_x,y_pr=False):
        if(y_pr==True):
            self.u=np.nanmean(train_x,axis=0)
            self.s=np.nanmax(train_x,axis=0)-np.nanmin(train_x,axis=0)+1
        train_x=(train_x-self.u)/self.s
        return train_x
    
    def normalizey(self,train_y,y_pr=False):
        if(y_pr==True):
            self.yu=np.nanmean(train_y,axis=0)
            self.ys=np.nanmax(train_y,axis=0)-np.nanmin(train_y,axis=0)+1
        train_y=(train_y-self.yu)/self.ys
        return train_y     
        
    def fit(self,train_x,train_y):
        train_x=np.nan_to_num(train_x)
        train_y.resize(train_y.shape[0],1)
        train_y=np.nan_to_num(train_y)
        b=np.ones((train_x.shape[0],1))
        train_x=np.hstack((b,train_x))
        train_x=self.normalizex(train_x,y_pr=True)
        train_y=self.normalizey(train_y,y_pr=True)
        #print(train_x.shape,train_y.shape)
        self.param_vec(train_x)
        #print(train_x.shape,train_x,self.param.shape)
        for i in range(self.iter):
            for j in range(0,train_x.shape[0],int(train_x.shape[0]/65)):
                self.predict(train_x[j:j+20][:])
            cost=self.cost_function(train_y[j:j+20][:])
            self.descent(train_x[j:j+20][:],train_y[j:j+20][:])
            if(i%100==0):
                print("Cost:"+str(cost))
    def predictions(self,X_test):
        X_test=np.nan_to_num(X_test)
        b=np.ones((X_test.shape[0],1))
        X_test=np.hstack((b,X_test))
        X_test=self.normalizex(X_test,y_pr=False)
        self.predict(X_test)
    def accuracy(self,X_test,Y_test):
        self.predictions(X_test)
        Y_test=Y_test.reshape(len(Y_test),1)
        Y_test=self.normalizey(Y_test)
        print("\nMean Absolute Error : ",MAE(Y_test,self.pred_y))
```

```python
# Importing Data
data=pd.read_csv('train.csv')
Y=data.loc[:,'SalePrice']
X=data.loc[:,['LotFrontage','LotArea','OverallQual','OverallCond','MasVnrArea']]
Y=Y.values
X=X.values
train_x=X[0:1300][:]
test_x=X[1300:][:]
train_y=Y[0:1300][:]
test_y=Y[1300:][:]
```

```python
obj=LinReg(9999999999,0.001)
obj.fit(train_x=train_x,train_y=train_y)
```

```python
obj.accuracy(test_x,test_y)
```
